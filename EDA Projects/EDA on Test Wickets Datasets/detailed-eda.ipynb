```python
counrty_dict = {"SL":"Sri Lanka",
        "AUS":"Australia",
        "ENG":"England",
        "INDIA":"India",
        "WI":"West Indis",
        "SA":"South Africa",
        "NZ":"Newziland",
        "PAK":"Pakisthan",
        "BAN":"Bangladesh",
        "ZIM":"Zimbabwe"}
```

```python
countries = project_data["Country"] .unique()
```

## Economy of each player

```python
for country in countries:
    df_country = project_data[project_data["Country"] == country]
    print(counrty_dict.get(country))
    fig = px.bar(data_frame=df_country.sort_values(by=["Economy"], ascending = False), x='Player',y='Economy', template='plotly_dark')
    fig.show()
```

## Number of Matches Played vs Wickets Taken

```python
for country in countries:
    df_country = project_data[project_data["Country"] == country]
    print(counrty_dict.get(country))
    player_names = df_country['Player']
    fig = go.Figure(data=[
        go.Bar(name='Matches', x=player_names, y=df_country['Matches']),
        go.Bar(name='Wickets', x=player_names, y=df_country['Wickets'])
    ])

    # Change the bar mode
    fig.update_layout(barmode='group', plot_bgcolor = 'black')
    fig.show()
```

## Runs given by each player

```python
for country in countries:
    df_country = project_data[project_data["Country"] == country]
    print(counrty_dict.get(country))
    fig = px.bar(data_frame=df_country.sort_values(by=["Runs"], ascending = False), x='Player',y='Runs', template='plotly_dark')
    fig.show()
```

## Matches vs Runs Analysis

```python
for country in countries:
    df_country = project_data[project_data["Country"] == country]
    print(counrty_dict.get(country))
    player_names = df_country['Player']
    fig = go.Figure(data=[
        go.Bar(name='Matches', x=player_names, y=df_country['Matches']),
        go.Bar(name='Runs', x=player_names, y=df_country['Runs'])
    ])

    # Change the bar mode
    fig.update_layout(barmode='group', plot_bgcolor = 'black')
    fig.show()
```

## Career Span Analysis

```python
for country in countries:
    df_country = project_data[project_data["Country"] == country]
    print(counrty_dict.get(country))
    fig = px.bar(data_frame=df_country.sort_values(by=["Career Span"], ascending = False), x='Player',y='Career Span', template='plotly_dark')
    fig.show()
```
